5)What is the average family income of the student?


```python
import numpy as np

# Define a function to convert income ranges to their midpoints
def convert_to_midpoint(income_range):
    try:
        # Extract numbers from the range and remove " Lakh"
        lower, upper = income_range.replace(' Lakh', '').split('-')
        # Convert to float and calculate midpoint
        midpoint = (float(lower) + float(upper)) / 2
        return midpoint
    except ValueError:
        # If the format is incorrect, return NaN
        return np.nan

# Apply the conversion function to the 'Family Income' column
data['Family Income Midpoint'] = data['Family Income'].apply(convert_to_midpoint)

# Calculate the average of the midpoint values
average_family_income = data['Family Income Midpoint'].mean()

print(f'Average family income of the students: {average_family_income:.2f} Lakh')
```

6)How does the GPA vary among different colleges? (Show top 5 results only)


```python
# Group by 'College Name' and calculate the average GPA for each college
college_gpa = data.groupby('College Name')['CGPA'].mean()

# Sort the colleges by average GPA in descending order and select the top 5
top_5_colleges = college_gpa.sort_values(ascending=False).head(5)

print("Top 5 Colleges by Average GPA:")
print(top_5_colleges)
```

7)Are there any outliers in the 'attending status & quantity (number of courses completed) attribute?


```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for 'Quantity'
Q1 = data['Quantity'].quantile(0.25)
Q3 = data['Quantity'].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define outliers as values that are below (Q1 - 1.5 * IQR) or above (Q3 + 1.5 * IQR)
outliers = data[(data['Quantity'] < (Q1 - 1.5 * IQR)) | (data['Quantity'] > (Q3 + 1.5 * IQR))]

print(f"Number of outliers: {outliers.shape[0]}")
print(outliers[['Quantity']].head())
```

8)What is the average GPA for student from each city?


```python
# Calculate the average GPA (CGPA) for each city
average_gpa_by_city = data.groupby('City')['CGPA'].mean().reset_index()

# Rename columns for better clarity
average_gpa_by_city.columns = ['City', 'Average CGPA']

print(average_gpa_by_city)
```

9)Can we identify any relationship between family income and GPA?

```python
import pandas as pd

# Load the dataset
data = pd.read_excel('data.xlsx')

# Map the 'Family Income' values to numeric categories
income_mapping = {
    '0-2 Lakh': 1,
    '2-5 Lakh': 2,
    '5-7 Lakh': 3,
    '7 Lakh+': 4
}
data['Family Income Encoded'] = data['Family Income'].map(income_mapping)

# Calculate the correlation between 'Family Income Encoded' and 'CGPA'
correlation = data[['Family Income Encoded', 'CGPA']].corr().iloc[0, 1]

# Output the correlation value
print(f"Correlation between Family Income and CGPA: {correlation}")
```
